Plotting sin linea.
para trazar solo los marcaores ,puedes usar notacion de cadena de acceso directo 'o',que significa "anillos".
ejemplos:dibuja dos puntos en el diagrama,uno en la pocision(1,3) y el otro en la posicion (8,10)

```python
import matplotlib.pyplot as plt
import numpy as np

xpuntos=np.array([1,8])
ypuntos=np.array([3,10])

plt.plot(xpuntos,ypuntos,'o')
plt.show()
```

diagramas de dispersion (scatter)

se emplean para mostrar la relacion entre dos variable numericas, dtectando correlacionales o agrupamientos .
ejemplo:


```python
import matplotlib.pyplot as plt

altura=[150,160,170,180,190,200]
peso=[50,60,65,80,85,95]
plt.scatter(altura,peso,color="red",marker="x")
plt.title("relacion entre la altura y el peso")
plt.xlabel("altura(cms)")
plt.ylabel("peso(kg)")
plt.show()
```

graficos de barras (bar y barh)
 sirven para comparar cstegorias o variables tanto en orientacion vertical como en horizontal.
 ejemplo de baaras verticales:


```python
import matplotlib.pyplot as plt
categorias=["a","b","c","d"]
valores=[4,7,1,8]

plt.bar(categorias,valores,color="orange")
plt.title("grafica de barras")
plt.xlabel("categorias")
plt.ylabel("valores")
plt.show()
```
